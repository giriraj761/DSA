Write a Python program to create a pie chart of the popularity of programming Languages.

```python

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['red', 'blue', 'green', 'cyan', 'pink', 'purple']
plt.pie(popularity,labels=languages, colors=colors, autopct='%1.1f%%')
legend = plt.legend(title='Programming Languages 1973', loc='lower left', bbox_to_anchor=(0.5, 0.9))
plt.show()
```

i. Write a Python program to plot two or more lines with legends, different widths and colors.

ii. Write a Python program to plot two or more lines with different styles.

iii. Write a Python program to plot two or more lines and set the line markers.

```python
#3 (I)
x = [1, 2, 3, 4, 5]
y1 = [3, 5, 6, 8, 9]
y2 = [1, 2, 4, 7, 10]
plt.plot(x, y1, color='blue', linewidth=2, label='Line 1')
plt.plot(x, y2, color='red', linewidth=3, label='Line 2')
plt.title('Two Lines with Different Colors and Widths GIRIRAJ 1973 ')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
# 3(II)
plt.plot(x, y1, linestyle='dashed', label='Line 1')
plt.plot(x, y2, linestyle='dotted', label='Line 2')
plt.title('Two Lines with Different Styles  GIRIRAJ 1973')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
#3(III)
plt.plot(x, y1, marker='o', label='Line 1')
plt.plot(x, y2, marker='s', label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Two Lines with Different Styles  GIRIRAJ 1973 GIRIRAJ 1973')
plt.show()
```

Calculate total sale data for last year for each product and show it using a Pie chart
