```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import math as m
import numpy as np
```

```python
#scatterplot 
data = pd.read_csv('Salaries.csv')
plt.scatter(data['rank'], data['salary'])
plt.title("Scatter Plot")
plt.show()
```

```python
#line scatter plot
plt.plot(data['salary'])
plt.show()
```

```python
#scatter plot
x=[1,2,3,4,5,6,7,8]
y=[i*i for i in x]
z = [m.log(x) for x in x]
plt.scatter(x,z)
# plt.scatter(x,y)
```
